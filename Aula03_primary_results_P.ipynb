## <font color = red> - Medidas Estatísticas
`np.mean` ou `st.tmean` ===> Média<br>
`np.median` ===> Mediana<br>
`st.mode` ===> Moda<br>
`np.var` ou `st.tvar` ===> Variância<br>
`np.std` ou `st.tstd` ===> Desvio Padrão<br>
`np.sum` ===> Soma<br>
`'count'` ===> Contar Valores<br>
`'nunique'` ===> Quantos Valores Diferentes tem<br>
`min` ou `st.tmin` ===> Valor Mínimo<br>
`max` ou `st.tmax` ===> Valor Máximo<br>
`np.ptp` ===> Amplitude<br>
`st.sem` ===> Erro Padrão<br>

 - Criar uma função para calcular o percentil


```python
def percentil(n):
    'Esta é uma função para calcular percentil agregado'
    def percentil_(x):
        return np.percentile(x, n)
    percentil_.__name__ = 'perc_%s' % n
    return percentil_
```

## <font color = red> # `Groupby`
- Faz um agrupamento entre duas ou mais variáveis

##### Referência
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html


```python
# Quer saber as medidas estatísticas dos candidatos em relação ao voto
#fazer um agrupamento dos candidatos em relação aos votos
df_cand = df.groupby('candidate').agg({'votes': [np.mean, np.median, st.mode, percentil(25), percentil(75),
                                                 np.var, np.std, np.sum, 'count', min, max, np.ptp, st.sem,
                                                 'nunique']})
```

```python
df_cand
```

- <font color = green> - Renomear as colunas

```python
df_cand = df_cand.rename(columns = {'mean': 'media', 'median': 'mediana', 'mode': 'moda','sum': 'soma'})
df_cand
```

```python
# Fazer um sort para saber que teve mais votos em média
df_cand.sort_values([('votes', 'media')], ascending = False)
```

- Agrupar por duas ou mais variáveis

```python
# Queremos saber a média dos votos por candidato e por estado
df_cand2 = df.groupby(['state', 'candidate']).agg({'votes': [np.mean]})
df_cand2
```
