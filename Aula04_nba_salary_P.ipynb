## <font color = red># `split`
- Separar por alguma referÃªncia
### ReferÃªncia
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.split.html

- Separar o nome do jogador da posiÃ§Ã£o que ele joga, para isso irÃ¡ se criar uma coluna 'POSITION'

```python
df[['NAME', 'POSITION']] = df.NAME.str.split(',', expand = True)
```

```python
df.head()
```

<font color = red>Fazer outra coluna com o Nome e o Sobre nome dos jogadores 

- Separar o nome do sobrenome do jogador<br>


```python
######################### -----  ERRO PROPOSITAL ------- #######################################33
#obs: Se repetir o mesmo cÃ³digo acima vai dar um erro de tamanho de chaves. As colunas devem ter o mesmo tamanho da chave

df[['NAME', 'L_NAME']] = df.NAME.str.split(' ', expand = True)
```

```python
'''
Para resolver o erro Ã© preciso acrescentar o n = 1 no comando split
'''

df[['NAME', 'L_NAME']] = df['NAME'].str.split(' ', n = 1, expand = True)
```

```python
df.head()
```

## <font color = red># `rename`
- Renomear coluna

### ReferÃªncia
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rename.html

- Renomear coluna 'NAME' para 'F_NAME'

```python
df = df.rename(index=str, columns={'NAME': 'F_NAME'})
```

```python
df.head()
```

- Colocar os nomes das colunas em caixa baixa

```python
df = df.rename(str.lower, axis='columns')
```

- Colocar os nomes das colunas em CAIXA ALTA

## <font color = red># `replace`

- Substituir ou Retirar Caracteres 

##### ReferÃªncia

- Substituir o `'$'` por `'#'` da variÃ¡vel SALARY

## <font color = red># PIVOT_TABLE
### ReferÃªncia
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html

- Deseja-se saber os SalÃ¡rio dos times por posiÃ§Ã£o

```python
pd.pivot_table(df, values='SALARY', index=['TEAM'], columns=['POSITION'], aggfunc = 'sum')
```

8 - Fazer um grÃ¡fico de salarios por time comparando por posiÃ§Ã£o<br>
Obs: colocar (col_wrap = 3) na linha de comando do grÃ¡fico

```python
# Compara os Candidatos dentro do Partido
g = sns.catplot(x = 'TEAM', y = 'SALARY',  data = df, col = 'POSITION',ci = None, kind = 'bar', col_wrap = 3)
g.set_xticklabels(rotation = 90)
plt.subplots_adjust(top = 0.8)
g.fig.suptitle('ComparaÃ§Ã£o dos SalÃ¡rios dos Times por PosiÃ§Ã£o', size  = 20, color = '#000000', weight = 'bold', style = 'italic')
fig.savefig('C:\\Users\\Erivelton\\python\\Curso de Estatistica\\Aula_Estat_Basica_python\\Exercicio_resp\\nba_salary_08.jpeg')
plt.show()
```
